```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession
import seaborn as sns
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MySparkApp") \
    .getOrCreate()

print("Spark version:", spark.version)
```

```python
df = spark.read.csv("House_Price_dataset.csv", header=True, inferSchema=True)
```

```python
df.show(5)
```

```python
df.createOrReplaceTempView("house_prices")
```

### OverView Analysis

```python
CityPriceAvg = spark.sql("""
    SELECT city, ROUND(AVG(price), 2) AS avg_price
    FROM house_prices
    GROUP BY city
    ORDER BY avg_price DESC
""")
CityPriceAvg.show()
```

```python
PropTypePriceAvg = spark.sql("""
    SELECT property_type,
           ROUND(AVG(price), 2) AS avg_price
    FROM house_prices
    GROUP BY property_type
    ORDER BY avg_price DESC
""")
PropTypePriceAvg.show()
```

### City - Price Average

Lahore

```python
LahorePriceAvg = spark.sql("""
    SELECT area, ROUND(AVG(price), 2) AS avg_price
    FROM house_prices
    WHERE city == "Lahore" 
    GROUP BY area                    
    ORDER BY avg_price DESC
""")
LahorePriceAvg.show()
```

### Same Area - Price Differences

```python
CityPricediff = spark.sql("""
    SELECT `Area Category`,
           ROUND(AVG(CASE WHEN city = 'Rawalpindi' THEN price END), 2) AS Rawalpindi,
           ROUND(AVG(CASE WHEN city = 'Lahore' THEN price END), 2) AS Lahore,
           ROUND(AVG(CASE WHEN city = 'Karachi' THEN price END), 2) AS Karachi,
           ROUND(AVG(CASE WHEN city = 'Islamabad' THEN price END), 2) AS Islamabad
    FROM house_prices
    GROUP BY `Area Category`
    ORDER BY `Area Category` ASC
""")

CityPricediff.show()
```
